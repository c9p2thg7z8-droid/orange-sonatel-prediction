```python
import pickle
import numpy as np

MODEL_PATH = "model_groupe.pkl"

# ─────────────────────────────────────────────
# Fonction de prédiction
# ─────────────────────────────────────────────
def predire_groupe(domaine, sous_domaine, type_, typologie):
    """
    Prédit le GROUPE pour une nouvelle plainte.

    Paramètres :
        domaine      : ex. 'MOBILE', 'FIXE', 'OM', 'DATA_MOBILE'
        sous_domaine : ex. 'RADIO_MOBILE', 'ADSL', 'VOIX_MOBILE'
        type_        : ex. 'PLAINTES GP', 'PLAINTES ENTREPRISE', 'PLAINTES VIP'
        typologie    : ex. 'COUVERTURE', 'NAVIGATION', 'CONNECTIVITE_OU_INSTABILITE'

    Retourne :
        (groupe prédit, score de confiance)
    """
    # Charger le modèle sauvegardé
    bundle   = pickle.load(open(MODEL_PATH, 'rb'))
    model    = bundle['model']
    encoders = bundle['encoders']
    features = bundle['features']   # ['DOMAINE', 'SOUS-DOMAINE', 'TYPE', 'TYPOLOGIE']

    # Préparer les valeurs dans le bon ordre
    vals = {
        'DOMAINE'      : domaine.strip().upper(),
        'SOUS-DOMAINE' : sous_domaine.strip().upper(),
        'TYPE'         : type_.strip().upper(),
        'TYPOLOGIE'    : typologie.strip().upper(),
    }

    # Encoder chaque feature (valeur inconnue → première classe connue par défaut)
    row = []
    for col in features:
        le = encoders[col]
        v  = vals[col]
        if v not in le.classes_:
            print(f"Valeur inconnue pour {col} : '{v}' → remplacée par '{le.classes_[0]}'")
            v = le.classes_[0]
        row.append(le.transform([v])[0])

    # Prédiction + confiance
    pred        = model.predict([row])[0]
    probas      = model.predict_proba([row])[0]
    confiance   = probas.max()
    groupe      = encoders['GROUPE'].inverse_transform([pred])[0]

    # Top 3 des groupes les plus probables
    top3_idx    = np.argsort(probas)[::-1][:3]
    top3        = [(encoders['GROUPE'].inverse_transform([i])[0], probas[i]) for i in top3_idx]

    print("\n" + "=" * 50)
    print("  RÉSULTAT DE PRÉDICTION")
    print("=" * 50)
    print(f"  Domaine      : {vals['DOMAINE']}")
    print(f"  Sous-domaine : {vals['SOUS-DOMAINE']}")
    print(f"  Type         : {vals['TYPE']}")
    print(f"  Typologie    : {vals['TYPOLOGIE']}")
    print("-" * 50)
    print(f" GROUPE prédit  : {groupe}")
    print(f" Confiance      : {confiance:.1%}")
    print("-" * 50)
    print("  Top 3 groupes possibles :")
    for i, (g, p) in enumerate(top3, 1):
        bar = "█" * int(p * 20)
        print(f"    {i}. {g:<30} {p:.1%}  {bar}")
    print("=" * 50)

    return groupe, confiance


# ─────────────────────────────────────────────
# TESTS - Modifie ces valeurs pour tester
# ─────────────────────────────────────────────
if __name__ == "__main__":

    # ── Test 1 : Problème de couverture mobile
    predire_groupe(
        domaine      = 'MOBILE',
        sous_domaine = 'RADIO_MOBILE',
        type_        = 'PLAINTES GP',
        typologie    = 'COUVERTURE'
        
    )

    # ── Test 2 : Problème ADSL
    predire_groupe(
        domaine      = 'FIXE',
        sous_domaine = 'ADSL',
        type_        = 'PLAINTES ENTREPRISE',
        typologie    = 'CONNECTIVITE_OU_INSTABILITE'
    )

    # ── Test 3 : Problème navigation data mobile
    predire_groupe(
        domaine      = 'DATA_MOBILE',
        sous_domaine = 'NAVIGATION',
        type_        = 'PLAINTES GP',
        typologie    = 'NAVIGATION'
    )

    # ── Test 4 : Voix fixe entreprise
    predire_groupe(
        domaine      = 'FIXE',
        sous_domaine = 'VOIX_FIXE',
        type_        = 'PLAINTES ENTREPRISE',
        typologie    = "EMISSION_ET_RECEPTION_D'APPEL"
    )
```
